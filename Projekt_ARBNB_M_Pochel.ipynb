```python
q1=dane['price'].quantile(0.25)
q3=dane['price'].quantile(0.75)

IQR=q3-q1
lower=q1-1.5*IQR
upper=q3+1.5*IQR

analysis=pd.DataFrame({'Q1': [q1], 'Q3':[q3], 'IQR': [IQR], 'Lower bound': [lower], "Upper bound": [upper]})
analysis
```

lower bound ustawiamy na 0 - nie ma ujemnych cen

```python
outliers=dane[(dane['price'] < 0) | (dane['price'] > upper)]
outliers
```

```python
q11=dane['reviews per month'].quantile(0.25)
q31=dane['reviews per month'].quantile(0.75)

IQR1=q31-q11
lower1=q11-1.5*IQR1
upper1=q31+1.5*IQR1

analysis1=pd.DataFrame({'Q1': [q11], 'Q3':[q31], 'IQR': [IQR1], 'Lower bound': [lower1], "Upper bound": [upper1]})
analysis1

outliers=dane[(dane['reviews per month'] < 0) | (dane['reviews per month'] > upper)]
outliers
```

```python

price_median = dane['price'].median()

dane['price_range'] = pd.cut(
    dane['price'],
    bins=[0, price_median, dane['price'].max()],
    labels=['low_price_segment', 'high_price_segment']
)

service_fee_summary = (
    dane.groupby('price_range')['service fee']
        .agg(['mean','median','count'])
        .round(2)
        .reset_index()
)
service_fee_summary

```

## Wpływ opcji instant bookable

Porównanie ofert z możliwością natychmiastowej rezerwacji oraz bez niej
nie wykazało istotnych różnic w cenie, liczbie recenzji ani ich ocenie.
Oznacza to, że funkcja instant bookable nie jest kluczowym czynnikiem
wpływającym na popyt ani cenę.


```python
dane['NAME']
```

```python
top1_percent=dane['price'].quantile(0.990)
bottom_1_prct=dane['price'].quantile(0.01)
top_expensive=dane[dane['price']>=top1_percent]
bottom_cheap=dane[dane['price']<=bottom_1_prct]
columns=['NAME',"price","neighbourhood","reviews per month", "review rate number","room type","calculated host listings count"]

top_rbnb=top_expensive[columns].sort_values(by='price')
bottom_listings_table = bottom_cheap[columns].sort_values(by='price')

```

## Analiza skrajnych ofert (top 1%)

Porównano 1% najdroższych oraz 1% najtańszych ofert.
Różnice pomiędzy tymi grupami dotyczą przede wszystkim:
- ceny
- typu pokoju
- liczby ofert posiadanych przez hosta


```python

top_rbnb['Type'] = 'Top 1% najdroższych'
bottom_listings_table['Type'] = 'Top 1% najtańszych'


final_df = pd.concat([top_rbnb, bottom_listings_table], ignore_index=True)


final_df = final_df.sort_values(by='price', ascending=False)

final_df = final_df.reset_index(drop=True)
final_df
```
